### Importação dos dados



```python
# Importando as bibliotecas necessárias para a análise dos dados
import pandas as pd
import matplotlib.pyplot as plt

# Carregando os dados das quatro lojas a partir dos arquivos CSV hospedados no GitHub
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
# Calculando o faturamento total de cada loja
faturamento_loja1 = loja["Preço"].sum()
faturamento_loja2 = loja2["Preço"].sum()
faturamento_loja3 = loja3["Preço"].sum()
faturamento_loja4 = loja4["Preço"].sum()

print(f'Faturamento Total Loja 1: R$ {faturamento_loja1}')
print(f'Faturamento Total Loja 2: R$ {faturamento_loja2}')
print(f'Faturamento Total Loja 3: R$ {faturamento_loja3}')
print(f'Faturamento Total Loja 4: R$ {faturamento_loja4}')

faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
```

```python
# Plotando o gráfico de barras
faturamento_Total = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
grafico = ['#FF5733', '#33FF57', '#3357FF', '#F333FF']

plt.figure(figsize=(10, 8))
bars = plt.bar(lojas,faturamento_Total, color = grafico, width = 0.3)
plt.xlabel('Lojas')
plt.ylabel('Faturamento em Reais')
plt.title('Faturamento por Loja', fontsize=14, fontweight='bold')
plt.bar_label(bars, fmt='R${:,.2f} Mi')
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
lista_lojas = [loja, loja2, loja3, loja4]
nomes_colunas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

def mostrar_analise_loja(dados_loja, nome):
    vendas = dados_loja['Produto'].value_counts()

    mais_vendidos = vendas.head(5).reset_index()
    mais_vendidos.columns = ['Produto', 'Número de Vendas']

    menos_vendidos = vendas.tail(5).reset_index()
    menos_vendidos.columns = ['Produto', 'Número de Vendas']

    print(f'{'='*30}')
    print(f'ANÁLISE DE VENDAS: {nome}')
    print(f'{'='*30}')

    print(f'\n PRODUTOS MAIS VENDIDOS:')
    print(mais_vendidos.to_string(index=False))

    print(f'\n PRODUTOS MENOS VENDIDOS:')
    print(menos_vendidos.to_string(index=False))
    print('\n')

    plt.figure(figsize = (10, 6))
    bars = plt.bar(mais_vendidos['Produto'], mais_vendidos['Número de Vendas'], color = 'green', width = 0.3)
    plt.xlabel('Produtos')
    plt.ylabel('Número de Vendas')
    plt.title(f'TOP 5 PRODUTOS MAIS VENDIDOS - {nome}', fontsize = 14, fontweight = 'bold')
    plt.bar_label(bars, fmt = '%d')
    plt.show()

    plt.figure(figsize = (10, 6))
    bars = plt.bar(menos_vendidos['Produto'], menos_vendidos['Número de Vendas'], color = 'red', width = 0.3)
    plt.xlabel('Produtos')
    plt.ylabel('Número de Vendas')
    plt.title(f'TOP 5 PRODUTOS MENOS VENDIDOS - {nome}', fontsize = 14, fontweight = 'bold')
    plt.bar_label(bars, fmt = '%d')
    plt.show()

    return mais_vendidos, menos_vendidos

for i in range(len(lista_lojas)):
    mostrar_analise_loja(lista_lojas[i], nomes_colunas[i])
```

# 5. Frete Médio por Loja

```python
media_frete = []

for i in range(len(lista_lojas)):
    media = lista_lojas[i]['Frete'].mean()
    media_frete.append({'Loja': nomes_colunas[i], 'Frete Médio': media})

df_frete = pd.DataFrame(media_frete)
df_frete['Frete Médio'] = df_frete['Frete Médio'].map('R$ {:.2f}'.format)
```

```python
# Lista que armazenará o frete médio de cada loja
media_frete = []

# Loop para calcular o frete médio
for i in range(len(lista_lojas)):
    #Calcula a média da coluna "Frete" da loja atual
    media = lista_lojas[i]['Frete'].mean()
    media_frete.append({'Loja': nomes_colunas[i], 'Frete Médio': media})

# Cria um DataFrame com os resultados
df_frete = pd.DataFrame(media_frete)
df_frete['Frete Médio'] = df_frete['Frete Médio'].map('R$ {:.2f}'.format)

# Plotando o gráfico
plt.figure(figsize = (10, 6))
bars = plt.bar(df_frete['Loja'], df_frete['Frete Médio'].str.replace('R$ ', '').astype(float), color = grafico, width = 0.3)
plt.xlabel('Lojas')
plt.ylabel('Frete Médio em Reais')
plt.title('Frete Médio por Loja', fontsize = 14, fontweight = 'bold')
plt.bar_label(bars, fmt = 'R$ {:.2f}')
plt.show()
```

# 6. Relatório Final

```python
melhor_faturamento = max(faturamentos)
melhor_faturamento_loja = lojas[faturamentos.index(melhor_faturamento)]
melhor_faturamento_loja
print(f'📈 Maior Faturamento: {melhor_faturamento_loja} - R${melhor_faturamento:.2f}')

melhor_avaliacao = max(avaliacoes_media)
melhor_avaliacao_loja = lojas[avaliacoes_media.index(melhor_avaliacao)]
melhor_avaliacao_loja
print(f'⭐ Maior Avaliação: {melhor_avaliacao_loja} - {round(melhor_avaliacao, 2)} de 5.0')

menor_frete_idx = df_frete['Frete Médio'].idxmin()
menor_frete_loja = df_frete.loc[menor_frete_idx, 'Loja']
menor_frete = df_frete.loc[menor_frete_idx, 'Frete Médio']
print(f'🚚+📉 Menor Frete: {menor_frete_loja} - R${menor_frete}')

total_vendas = sum(len(df) for df in lista_lojas)
soma_faturamentos = sum(faturamento_total for faturamento_total in faturamentos)
print('\n')
print(f'DADOS GERAIS:')
print(f'📊 Total de Vendas: {total_vendas} transações')
print(f'💵 Faturamento Total: R${soma_faturamentos}')
print(f'Receita Média por Venda: R${round((soma_faturamentos / total_vendas), 2)}')
print('\nAnálise Concluída!')
```
